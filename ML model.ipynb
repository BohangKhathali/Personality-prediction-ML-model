## STAGE 1: DATA CLEANING

```python
import pandas as pd
```

```python
df=pd.read_csv("mbti_1.csv")
```

```python
df
```

```python
# We will first clean the dataset by dropping any missing rows(if there arre any)
df=df.dropna(how="any")
```

```python
# We will remove any links to make it easier for our model to input the data. 
# These links don't tell us much about the person's personality to begin with aswell
df['posts'] = df['posts'].str.replace(r'http\S+', '', regex=True)
df
```

```python
# We want to simplify the types. Instead of focusing on 16 different personality types, 
# we will just create a new column and generalise these 16 personality types into two types 
# Introvert (I) vs Extrovert(E))
df["personality_type"]=df["type"].str[0]
df
```

```python
df=df[["posts","type","personality_type"]]
df
```

```python
# We now need to remove a records that mention any of the 16 persoanltity types (infj, entp and etc) and other cheat words
# Sometimes people will give away their personality type when typing, this is a cheat and will reduce our models accuracy.
df=df.copy()
mbti_types=["istj", "isfj", "infj", "intj", "istp", "isfp", 
            "infp", "intp", "estp", "esfp", "enfp", "entp", "estj", 
            "esfj", "enfj", "entj",  "introvert", "extrovert", "introversion", "extroversion", "introverted", "extroverted"]

df["cleaned_posts"]=df["posts"].str.lower()


for word in mbti_types:
    df["cleaned_posts"]=df["cleaned_posts"].str.replace(word, " ")

df=df[["posts","cleaned_posts","type","personality_type"]]
df
```

```python
# We will now remove the pipe operators(|||) that seperate the diffeerent entires someone made.
# We do this so that our ML model will be able to be more accurate and have less noise to work with

df["cleaned_posts"]= df["cleaned_posts"].str.replace("|||", " ", regex=False)
df
```

```python
# My dataset is now cleaned, I save it as a csv to to work on for the EDA?Visualisations part
df.to_csv("cleaned_data.csv", index=False)
```

## Phase 2: ML model

```python
# Since my ML model can only understand numbers, we will start by converting our personality types to numbers (0 and 1)
df["encoded_type"]=df["personality_type"].replace({"I":0,"E":1})
```

```python
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Here we transform our cleaned_posts column to a matrix of word importance scores using tf-idf vectorization
# the stop_words takes out all the common and boring English  words that people may use when typing like "the", "and", "this" and etc,
# We take out these words to reduce noise for our model
# max_features stops the model from getting distracted or misused from finding unecasary/rare words

tfidf=TfidfVectorizer(stop_words="english", max_features=5000)
X=tfidf.fit_transform(df["cleaned_posts"])
y=df["encoded_type"]
```

```python
from sklearn.model_selection import train_test_split

#Here we are splitting our data into training and testing
# 20% of the data will be used to test while 80% of the data will be used to train
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=42, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
# Now we create our model and train it
# We set our max_iter to 1000 so that the model has 1000 chances to try predict/guess the correct personality type before it has to quit and move on to the next person.
# Made class_weight balanced so that our model places equal importance amongst introverts and extroverts(the minority in this dataset)
model = LogisticRegression(max_iter=1000, class_weight="balanced")

model.fit(X_train, y_train)

print("Model training complete!")
```

```python
from sklearn.metrics import accuracy_score, classification_report
# We now use our model to predict the our testing data
y_pred = model.predict(X_test)

# We then get the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy * 100:.2f}%")

# Then we get the classification report of the model
print(classification_report(y_test, y_pred))

```

## User response

```python
# We ask for user input (our model will then predict what personality type the user is)
user_response = input("After a long, stressful week, how do you prefer to spend your Saturday evening to feel like 'yourself' again, and why?")


tfidf_person = tfidf.transform([user_response])

# We predict the users personality type
result = model.predict(tfidf_person)

# We then work out the probability of a users irtoversion vs extroversion
probability= model.predict_proba(tfidf_person)
introvert_chance = probability[0][0]
extrovert_chance = probability[0][1]

if result[0]==1:
    print("Our model predicts that this person is an Extrovert.", "Confidence level: ", round((extrovert_chance*100),2),"%")
else:
    
    print("Our model predicts that this person is an Introvert.","Confidence level: ", round((introvert_chance*100),2),"%")
```

## Exporting model results and info/report

```python
# Here we will create a dataframe to save the model's results for the dataset's (MBTI) tested data (that 20% test data).
y_probability=model.predict_proba(X_test)
model_results=pd.DataFrame({"Actual_type": y_test, 
                            "Predicted_type":y_pred, 
                            "Confidence_score": [max(prob) for prob in y_probability]}) 

# We then exort this as a csv
model_results.to_csv("people_results.csv", index=False)
```
